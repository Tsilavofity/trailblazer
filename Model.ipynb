```python
# Drop column target
y_train = train['target']
train.drop('target', axis = 1, inplace = True)
y_train
```

```python
X_train = train.drop(['ward', 'ADM4_PCODE'], axis = 1)
X_test = test.drop(['ward', 'ADM4_PCODE'], axis = 1)
X_train
```

## Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
num = X_train
num_test = X_test
num.columns
```

```python
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train,columns=num.columns,index=num.index)
X_test=pd.DataFrame(X_test,columns=num_test.columns,index=num_test.index)
X_test
```

```python
X_train['ward'] = train['ward']
X_test['ward'] = test['ward']
X_train['ADM4_PCODE'] = train['ADM4_PCODE']
X_test['ADM4_PCODE'] = test['ADM4_PCODE']
X_test
```

```python
cols = train.columns
X_train = X_train.reindex(train.columns, axis=1)
cols = test.columns
X_test = X_test.reindex(test.columns, axis=1)
```

## Model selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
def evaluate(model):
    model.fit(X_train.drop(['ward', 'ADM4_PCODE'], axis = 1),y_train) 
    pred = model.predict(X_train.drop(['ward', 'ADM4_PCODE'], axis = 1))
    score = r2_score(y_train,pred)
    print('r2 score = ',score)
    print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_train,pred)))
    print("Cross validation score", np.sqrt(-np.mean(cross_val_score(model, X_train.drop(['ward', 'ADM4_PCODE'], axis = 1), y_train, cv=5,scoring="neg_root_mean_squared_error"))))
```

```python
# Linear regression
Linear = LinearRegression()
evaluate(Linear)
```

```python
# Random forest
RF = RandomForestRegressor(n_estimators=187, max_features=15)
evaluate(RF)
```

```python
# XGBoost
XGB = XGBRegressor(n_estimators=187, learning_rate=0.05)
evaluate(XGB)
```

```python
# CatBoost
cat = CatBoostRegressor()
evaluate(cat)
```
